```python
###
#   Darwin Liao. Shopify Data Science Challenge 2022
#
#
#initilization
import numpy as np # linear algebra
import pandas as pd # data processing, file I/O (e.g. pd.read_csv)
import sklearn     #Used for simple machine learning algorithms

#Reading in the file
filename = "2019 Winter Data Science Intern Challenge Data Set.xlsx"
data = pd.read_excel(filename)
data.head()
```

```python
#Finding out and setting up avg and limits for the values
avgOrdAmt = data["order_amount"].sum()/data["order_amount"].count()
avgItmAmt = data["total_items"].sum()/data["total_items"].count()
ordLim = avgOrdAmt * 2
itmLim = avgItmAmt * 2
print("Self Imposed Order Limit: " + str(ordLim))
print("Self Imposed Amount Limit: " + str(itmLim))
```

```python
#make an "is_outlier" column for future AI to determine if it is an outlier or not
data["is_outlier"] = np.nan
data.head()
```

```python
#Fill in the Outlier based on previous conditions.
data["is_outlier"] = np.where((data["order_amount"] > ordLim)|(data["total_items"] > itmLim),True,False)
data.head()
```
